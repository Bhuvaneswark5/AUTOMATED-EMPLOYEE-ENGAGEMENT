# TASK 1

Write a program to develop a simple console-based computer brain teaser, a riddle logic puzzle, text-based puzzles involving wordplay, logic, and lateral thinking games in Python. The program will randomly choose a riddle question from a list of riddles (question-answer dictionary) provided in the program. It will then prompt the user for an answer and compare it against the correct answer.
If the user's answer matches the correct one, the program will print 'Correct'. Otherwise, it will print 'Incorrect' and provide the correct answer. The program will continue to run infinitely, allowing users to solve multiple riddles with the option to play again. Upon exiting the game, the program should display the user's score as a percentage

```python
import random

# Dictionary of riddles and their corresponding answers
riddles = {
    "I speak without a mouth and hear without ears. I have no body, but I come alive with the wind. What am I?": "echo",
    "The more of this there is, the less you see. What is it?": "darkness",
    "What comes once in a minute, twice in a moment, but never in a thousand years?": "the letter 'm'",
    "What has keys but can't open locks?": "piano",
    "What has many keys but can't open a single lock?": "keyboard"
}

def riddle_game():
    score = 0
    total_riddles = len(riddles)
    riddle_keys = list(riddles.keys())
    
    while True:
        # Randomly select a riddle
        riddle = random.choice(riddle_keys)
        correct_answer = riddles[riddle]
        
        # Display the riddle to the user
        print("\nRiddle:")
        print(riddle)
        
        # Prompt user for an answer
        user_answer = input("Your answer: ").strip().lower()
        
        # Check if the answer is correct
        if user_answer == correct_answer:
            print("Correct!")
            score += 1
        else:
            print(f"Incorrect. The correct answer is '{correct_answer}'.")
        
        # Ask if the user wants to play again
        play_again = input("\nDo you want to play again? (yes/no): ").strip().lower()
        if play_again != 'yes':
            break
    
    # Calculate and display the final score as a percentage
    if total_riddles > 0:
        percentage_score = (score / total_riddles) * 100
    else:
        percentage_score = 0
    
    print(f"\nYour final score: {percentage_score:.2f}%")

# Entry point of the program
if __name__ == "__main__":
    print("Welcome to the Riddle Game!")
    riddle_game()
```

# TASK 2

Write a program to develop a simple console-based computer brain teaser, an arithmetic logic puzzle, a problem-solving game in Python. The program will randomly generate two numeric numbers within the range of 1 to 20 and an arithmetic operator ('+', '-', 'P', '*').
The program will run infinite turns to solve the arithmetic puzzle with the option to play again. 13

The program should handle exceptions gracefully, such as invalid inputs when the user attempts to solve the puzzle. The program should print the percentage of the score when the user leaves the game

```python
import random

def generate_problem():           #Generate a random arithmetic problem
    num1 = random.randint(1, 20)
    num2 = random.randint(1, 20)
    operator = random.choice(['+', '-', '*', '/'])
    
   
    if operator == '/' and num2 == 0:   # Ensure the division operator (/) generates a valid problem
        num2 = random.randint(1, 20)   # Avoid division by zero
    
    problem = f"{num1} {operator} {num2}"
    return problem

def solve_problem(problem):
    """Evaluate the arithmetic expression and return the result."""
    try:
        result = eval(problem)
        return result
    except ZeroDivisionError:
        return None
    except SyntaxError:
        return None
    except TypeError:
        return None

def main():
    print("Welcome to the Arithmetic Puzzle Game!")
    correct_count = 0
    total_count = 0
    
    while True:
        problem = generate_problem()
        print(f"Calculate: {problem}")
        
        try:
            user_answer = input("Your answer (or 'q' to quit): ").strip().lower()
            
            if user_answer == 'q':
                break
            
            total_count += 1
            correct_answer = solve_problem(problem)
            
            if correct_answer is None:
                print("Invalid operation! Please try again.")
                continue
            
            try:
                user_answer = float(user_answer)
                if user_answer == correct_answer:
                    print("Correct!")
                    correct_count += 1
                else:
                    print(f"Incorrect. The correct answer is {correct_answer}.")
            except ValueError:
                print("Invalid input! Please enter a number or 'q' to quit.")
        
        except KeyboardInterrupt:
            print("\n\nExiting game...")
            break
    
    if total_count > 0:
        score_percentage = (correct_count / total_count) * 100
        print(f"\nYour score: {correct_count}/{total_count} ({score_percentage:.2f}% correct)")
    else:
        print("\nNo games played. Goodbye!")

if __name__ == "__main__":
    main()
```

```python
import random

# Define questions in a dictionary with keys as category names
questions = {
    'Science': [
        {"question": "What is the chemical symbol for potassium?", "answer": "k"},
        {"question": "What is the name of 1st element in perodic table?", "answer": "hydrogen"},
        {"question": "What is the powerhouse of the cell?", "answer": "mitochondria"},
    ],
    'Mathematics': [
        {"question": "What is the square root of 100?", "answer": "10"},
        {"question": "What is the sum 100 and 200", "answer": "300"},
        {"question": "What is the value of Pi (Ï€) to two decimal places?", "answer": "3.14"},
    ],
    'Geography': [
        {"question": "What is the capital city of india?", "answer": "new delhi"},
        {"question": "Which state has highest literacy rate", "answer": "kerala"},
        {"question": "What is the longest river in the world?", "answer": "amazon river"},
    ],
    'Other': [
        {"question": "Capital city of bangladesh", "answer": "dhaka"},
        {"question": "What is the largest mammal on Earth?", "answer": "blue whale"},
        {"question": "Which country is famous for Mount Everest?", "answer": "nepal"},
    ]
}

def ask_category():
    """Ask the user to choose a category."""
    print("Select a category:")
    for index, category in enumerate(questions.keys(), 1):
        print(f"{index}. {category}")
    choice = input("Enter your choice (1-4): ").strip()
    return choice

def present_category_options():
    """Present category options and return user's choice."""
    choice = ask_category()
    while choice not in ['1', '2', '3', '4']:
        print("Invalid choice. Please enter a number between 1 and 4.")
        choice = ask_category()
    return choice

def ask_question(category):
    """Select and return a random question from the specified category."""
    category_name = list(questions.keys())[int(category) - 1]
    question = random.choice(questions[category_name])
    return question

def display_question(question):
    """Display the given question."""
    print("\nQuestion:", question["question"])

def prompt_user_answer():
    """Prompt the user for an answer."""
    return input("Your answer: ").strip().lower()

def check_answer(user_answer, correct_answer):
    """Compare the user's answer with the correct answer."""
    return user_answer == correct_answer

def play_game():
    """Manages the main game loop."""
    print("Welcome to the General Knowledge Quiz Game!")
    play_again = True
    total_questions = 0
    correct_answers = 0
    
    while play_again:
        category_choice = present_category_options()
        question = ask_question(category_choice)
        
        display_question(question)
        user_answer = prompt_user_answer()
        
        total_questions += 1
        if check_answer(user_answer, question["answer"]):
            print("Correct!")
            correct_answers += 1
        else:
            print(f"Incorrect. The correct answer is '{question['answer']}'.")

        play_choice = input("\nDo you want to play again? (yes/no): ").strip().lower()
        if play_choice != "yes":
            play_again = False
    
    print("\nQuiz Summary:")
    print(f"Total questions: {total_questions}")
    print(f"Correct answers: {correct_answers}")
    
    if total_questions > 0:
        percentage_correct = (correct_answers / total_questions) * 100
    else:
        percentage_correct = 0
    
    print(f"Percentage correct: {percentage_correct:.2f}%")
    print("\nThank you for playing!")

if __name__ == "__main__":
    play_game()
```

# Task 4

Think about how the rules of the Rock-Paper-Scissors game can be translated into Python code. Build the game using a Python program where Player 1 is the computer, making random choices.

```python
import random

def get_user_choice():
    """
    Function to get and validate user choice for Rock, Paper, or Scissors.
    """
    while True:
        user_choice = input("Enter your choice (rock/paper/scissors): ").lower()
        if user_choice in ['rock', 'paper', 'scissors']:
            return user_choice
        else:
            print("Invalid choice! Please enter rock, paper, or scissors.")

def get_computer_choice():
    """
    Function to get computer's random choice for Rock, Paper, or Scissors.
    """
    choices = ['rock', 'paper', 'scissors']
    computer_choice = random.choice(choices)
    return computer_choice

def determine_winner(user_choice, computer_choice):
    """
    Function to determine the winner of the game.
    """
    if user_choice == computer_choice:
        return "It's a tie!"
    elif (user_choice == 'rock' and computer_choice == 'scissors') or \
         (user_choice == 'paper' and computer_choice == 'rock') or \
         (user_choice == 'scissors' and computer_choice == 'paper'):
        return "You win!"
    else:
        return "Computer wins!"

def rock_paper_scissors():
    """
    Main function to play Rock-Paper-Scissors game.
    """
    print("Welcome to Rock-Paper-Scissors!")
    
    while True:
        user_choice = get_user_choice()
        computer_choice = get_computer_choice()
        
        print(f"\nYou chose: {user_choice}")
        print(f"Computer chose: {computer_choice}")
        
        result = determine_winner(user_choice, computer_choice)
        print(result)
        
        play_again = input("\nDo you want to play again? (yes/no): ").lower()
        if play_again != 'yes':
            print("Thanks for playing!")
            break

# Start the game
rock_paper_scissors()
```

# TASK 5

Develop a Python program to simulate a two-player Rock- Paper-Scissors game. The program should prompt each player to enter their name, allow each player to choose rock, paper, or scissors, implement logic to determine the winner, and display the winner's name

```python
def determine_winner(player1_name, player1_choice, player2_name, player2_choice):
    """
    Function to determine the winner of Rock-Paper-Scissors game.
    Returns the name of the winner, or None if it's a tie.
    """
    if player1_choice == player2_choice:
        return None  # It's a tie

    if (player1_choice == 'rock' and player2_choice == 'scissors') or \
       (player1_choice == 'paper' and player2_choice == 'rock') or \
       (player1_choice == 'scissors' and player2_choice == 'paper'):
        return player1_name  # Player 1 wins

    return player2_name  # Player 2 wins


def main():
    print("Welcome to the Rock-Paper-Scissors game!")

    # Get player names
    player1_name = input("Player 1, enter your name: ")
    player2_name = input("Player 2, enter your name: ")

    # Get player choices
    player1_choice = input(f"{player1_name}, choose rock, paper, or scissors: ").strip().lower()
    player2_choice = input(f"{player2_name}, choose rock, paper, or scissors: ").strip().lower()

    # Validate choices
    choices = ['rock', 'paper', 'scissors']
    while player1_choice not in choices or player2_choice not in choices:
        print("Invalid choice. Please choose rock, paper, or scissors.")
        player1_choice = input(f"{player1_name}, choose rock, paper, or scissors: ").strip().lower()
        player2_choice = input(f"{player2_name}, choose rock, paper, or scissors: ").strip().lower()

    # Determine the winner
    winner = determine_winner(player1_name, player1_choice, player2_name, player2_choice)

    # Print the result
    if winner is None:
        print("It's a tie!")
    else:
        print(f"The winner is: {winner}!")

if __name__ == "__main__":
    main()
```
